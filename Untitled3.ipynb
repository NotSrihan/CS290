```python
def plot_gaussian_mixture(clusterer, X, resolution=1000, show_ylabels=True):
    # Define grid boundaries
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))

    # Compute the negative log-probability (for contour plot)
    Z = -clusterer.score_samples(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot probability density contours
    plt.contourf(xx, yy, Z,
                 norm=LogNorm(vmin=1.0, vmax=30.0),
                 levels=np.logspace(0, 2, 12),
                 cmap="viridis")
    plt.contour(xx, yy, Z,
                norm=LogNorm(vmin=1.0, vmax=30.0),
                levels=np.logspace(0, 2, 12),
                linewidths=1, colors='k')

    # Plot decision boundaries
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z,
                linewidths=2, colors='r', linestyles='dashed')

    # Plot data points
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

    # Plot cluster centers
    plt.scatter(clusterer.means_[:, 0], clusterer.means_[:, 1],
                marker='X', s=200, c='white', edgecolor='k', label='Cluster Centers')

    # Label axes
    plt.xlabel("$x_1$")
    if show_ylabels:
        plt.ylabel("$x_2$", rotation=0)
    else:
        plt.tick_params(labelleft=False)

# Load Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use the first two features for simplicity

# Fit GaussianMixture model
gm = GaussianMixture(n_components=3, n_init=10, random_state=42)
gm.fit(X)

# Plot the Gaussian mixture
plt.figure(figsize=(8, 4))
plot_gaussian_mixture(gm, X)

# Show the plot
plt.legend()
plt.title("Gaussian Mixture Model with Decision Boundaries")
plt.show()
```
